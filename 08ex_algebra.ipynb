1\. **PCA on 3D dataset**

* Generate a dataset simulating 3 features, each with N entries (N being ${\cal O}(1000)$). Each feature is made by random numbers generated according the normal distribution $N(\mu,\sigma)$ with mean $\mu_i$ and standard deviation $\sigma_i$, with $i=1, 2, 3$. Generate the 3 variables $x_{i}$ such that:
    * $x_1$ is distributed as $N(0,1)$
    * $x_2$ is distributed as $x_1+N(0,3)$
    * $x_3$ is given by $2x_1+x_2$
* Find the eigenvectors and eigenvalues using the eigendecomposition of the covariance matrix
* Find the eigenvectors and eigenvalues using the SVD. Check that the two procedures yield to same result
* What percent of the total dataset's variability is explained by the principal components? Given how the dataset was constructed, do these make sense? Reduce the dimensionality of the system so that at least 99% of the total variability is retained
* Redefine the data according to the new basis from the PCA
* Plot the data, in both the original and the new basis. The figure should have 2 rows (the original and the new basis) and 3 columns (the $[x_0, x_1]$, $[x_0, x_2]$ and $[x_1, x_2]$ projections) of scatter plots.

```python
import numpy as np
from scipy import linalg as la
from matplotlib import pyplot as plt
import math

N=1000

x1=np.random.normal(0,1,N)
x2=x1+np.random.normal(0,3,N)
x3=2*x1+x2

M=np.array([x1,x2,x3])
cov=np.cov(M)

val, V=la.eig(cov)
val=np.real_if_close(val)


U, s, Vt = la.svd(M)

print("Eiginevalues:\n", val, '\n')
print("Eiginevectors:\n", V, '\n')
print("Eigenvalues with SVD: \n", s**2/(N-1),'\n')
print("Eigenvectors with SVD \n", U, '\n')
l=np.diag(val)
pca=0
sum=0
for i in range(len(val)):
    pca= l[i, i]/l.trace()
    sum=sum+pca
    print("By selecting the component ",i,":",pca)
    print("Sum of the % of variability explained by the components:",sum)
#The component x2 is almost useless, as all tha information are contained in x1 and x3
l_d=np.delete(l,1,1)
l_d=np.delete(l_d,1,0)
print(l_d)
print("% of variability explained by the components after dimension reduction:",(l_d[1, 1]+l_d[0,0])/l.trace())

Mnew=np.dot(V,M)

fig,ax=plt.subplots(figsize=(20, 10),nrows=2,ncols=3)
ax[0][0].scatter(x=M[0,:], y=M[1,:], marker='o', c='r', edgecolor='b')
ax[0][0].set_xlabel('$x1$')
ax[0][0].set_ylabel('$x2$')
ax[0][1].scatter(x=M[0,:], y=M[2,:], marker='o', c='r', edgecolor='b')
ax[0][1].set_xlabel('$x1$')
ax[0][1].set_ylabel('$x3$')
ax[0][2].scatter(x=M[1,:], y=M[2,:], marker='o', c='r', edgecolor='b')
ax[0][2].set_xlabel('$x2$')
ax[0][2].set_ylabel('$x3$')
ax[1][0].scatter(x=Mnew[0,:], y=Mnew[1,:], marker='o', c='r', edgecolor='b')
ax[1][0].set_xlabel('$x1new$')
ax[1][0].set_ylabel('$x2new$')
ax[1][1].scatter(x=Mnew[0,:], y=Mnew[2,:], marker='o', c='r', edgecolor='b')
ax[1][1].set_xlabel('$x1new$')
ax[1][1].set_ylabel('$x3new$')
ax[1][2].scatter(x=Mnew[1,:], y=Mnew[2,:], marker='o', c='r', edgecolor='b')
ax[1][2].set_xlabel('$x2new$')
ax[1][2].set_ylabel('$x3new$')
```

2\. **PCA on a nD dataset**

* Start from the dataset you have genereted in the previous exercise and add uncorrelated random noise. Such noise should be represented by other 10 uncorrelated variables normally distributed, with a standard deviation much smaller (e.g. a factor 20) than those used to generate the $x_1$ and $x_2$. Repeat the PCA procedure and compare the results with what you have obtained before.

```python
for i in range(3):
    for j in range(10):
        M[i,:]+=np.random.normal(0,1/10,N)
cov=np.cov(M)      
val, V=la.eig(cov)
val=np.real_if_close(val)
l=np.diag(val)
pca=0
sum=0
for i in range(len(val)):
    pca= l[i, i]/l.trace()
    sum=sum+pca
    print("By selecting the component ",i,":",pca)
    print("Sum of the % of variability explained by the components:",sum)
    
l_d=np.delete(l,1,1)
l_d=np.delete(l_d,1,0)
#now the pca procedure shows that x2 is not negligible, it shows about a 1% of the total data variability

Mnew=np.dot(V,M)

fig,ax=plt.subplots(figsize=(20, 10),nrows=2,ncols=3)
ax[0][0].scatter(x=M[0,:], y=M[1,:], marker='o', c='r', edgecolor='b')
ax[0][0].set_xlabel('$x1$')
ax[0][0].set_ylabel('$x2$')
ax[0][1].scatter(x=M[0,:], y=M[2,:], marker='o', c='r', edgecolor='b')
ax[0][1].set_xlabel('$x1$')
ax[0][1].set_ylabel('$x3$')
ax[0][2].scatter(x=M[1,:], y=M[2,:], marker='o', c='r', edgecolor='b')
ax[0][2].set_xlabel('$x2$')
ax[0][2].set_ylabel('$x3$')
ax[1][0].scatter(x=Mnew[0,:], y=Mnew[1,:], marker='o', c='r', edgecolor='b')
ax[1][0].set_xlabel('$x1new$')
ax[1][0].set_ylabel('$x2new$')
ax[1][1].scatter(x=Mnew[0,:], y=Mnew[2,:], marker='o', c='r', edgecolor='b')
ax[1][1].set_xlabel('$x1new$')
ax[1][1].set_ylabel('$x3new$')
ax[1][2].scatter(x=Mnew[1,:], y=Mnew[2,:], marker='o', c='r', edgecolor='b')
ax[1][2].set_xlabel('$x2new$')
ax[1][2].set_ylabel('$x3new$')
```
